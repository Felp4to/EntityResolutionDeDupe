```python
import json
import os
```

```python
PATH = "./Cleaned Datasets/"
PATH2 = "./Mediated Datasets/"
```

```python
# array containing fields of the mediated schema   
MEDIATED_STRUCTURE = [
    "name",
    "headquarters",
    "country",
    "continent",
    "region",
    "sub_region",
    "founded",
    "employees",
    "ceo",
    "market_cap",
    "categories",
    "sector",
    "industry",
    "revenue",
    "link",
    "share_price",
    "results_for_year",
    "total_equity",
    "founders",
    "market_value",
    "address",
    "type"
]

# match attributes
MATCH_ATTRIBUTES = {
    'name': 'name',
    #'BRAND NAME': 'name',
    'company': 'name',
    #
    'country': 'country',

    'headquarters_country': 'country',
    'location': 'country',
    #
    'sector': 'sector',
    'company_type': 'sector',

    #
    'industry': 'industry',
    #
    'founded': 'founded',
    'founded_year': 'founded',
    'ipo_year': 'founded',
    'founding_year': 'founded',
    'registration_date': 'founded',
    'family_ownership': 'founded',
    #
    'employees': 'employees',
    'number_of_employees': 'employees',
    #
    'ceo': 'ceo',
    #
    'market_capitalization_usd': 'market_cap',
    'market_cap': 'market_cap',
    'marketcap': 'market_cap',
    'mastercap': 'market_cap',
    'pricecap': 'market_cap',
    'master_cap': 'market_cap',
    'market_capitalization_2022': 'market_cap',
    'market_value_apr_2022': 'market_cap',
    #
    'share_price': 'share_price',
    #
    'revenue': 'revenue',
    'annual_revenue_in_usd': 'revenue',
    #
    #
    'link': 'link',
    'company_website': 'link',
    #
    'headquarters_continent': 'continent',
    'continent': 'continent',
    #
    'headquarters': 'headquarters',
    'head_quarters': 'headquarters',
    #
    'categories': 'categories',
    'company_business': 'categories',
    'category': 'categories',
    #
    'annual_results_for_year_ending': 'results_for_year',
    #
    'headquarters_region_city': 'region',
    #
    'headquarters_sub_region': 'sub_region',
    #
    'founders': 'founders',
    #
    'address': 'address',
    'registered_office_address': 'address',
    'office_address': 'address',
    #
    'public/private': 'type',
    'type': 'type'
}
```

```python
# return json with relabeling
def relabeling_projection_json(company):
    r_json = {}
    for field in company:
        if (field in MATCH_ATTRIBUTES) and (MATCH_ATTRIBUTES[field] in MEDIATED_STRUCTURE):
            r_json[MATCH_ATTRIBUTES[field]] = company[field]
    return json.dumps(r_json)


def relabeling_projection_file(fileName):
    file = open(PATH + fileName, 'r', encoding="utf8")
    rp_file = open(PATH2 + fileName.replace('.jsonl', '') + "_m" + ".jsonl", "w", encoding="utf8")
    json_str = ''
    for line in file:
        if line != '\n':
            json_str += relabeling_projection_json(json.loads(line)) + '\n'
    print(json_str, file=rp_file)
    rp_file.close()
    file.close()


# return files list
def find_all_files(path_dir):
    # list to store files
    res = []
    # Iterate directory
    for path in os.listdir(path_dir):
        # check if current path is a file
        if os.path.isfile(os.path.join(path_dir, path)):
            res.append(path)
    return res


# find folders list
def find_all_folders(path):
    return os.listdir(path)


# create folder
def create_folder(name, path):
    dir = os.path.join(path, name)
    if not os.path.exists(dir):
        os.mkdir(dir)


def relabeling_and_projection():
    files = find_all_files(PATH)
    for fileName in files:
        if fileName != '.DS_Store':
            relabeling_projection_file(fileName)
```
