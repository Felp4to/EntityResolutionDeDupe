```python
# return json with relabeling
def relabeling_projection_json(company):
    r_json = {}
    for field in company:
        if((field in MATCH_ATTRIBUTES) and (MATCH_ATTRIBUTES[field] in MEDIATED_STRUCTURE)):
            r_json[MATCH_ATTRIBUTES[field]] = company[field]
    return r_json

def relabeling_projection_file(fileName, source):
    file = open(PATH + source + "_original" + "/" + fileName, 'r', encoding="utf8")
    rp_file = open(PATH2 + source + "_mediated" + "/" + fileName.replace('.jsonl', '') + "_m" + ".jsonl", "w", encoding="utf8")
    for line in file:
        rp_file.write(json.dumps(relabeling_projection_json(json.loads(line))))
        rp_file.write('\n')
    rp_file.close()
    file.close()

# return files list
def find_all_files(path_dir):
    # list to store files
    res = []
    # Iterate directory
    for path in os.listdir(path_dir):
        # check if current path is a file
        if os.path.isfile(os.path.join(path_dir, path)):
            res.append(path)
    return res

# find folders list 
def find_all_folders(path):
    return os.listdir(path)

# create folder
def create_folder(name, path):
    dir = os.path.join(path, name)
    if not os.path.exists(dir):
        os.mkdir(dir)

        
def relabeling_and_projection():
    folders = find_all_folders(PATH)
    for folder in folders:
        label = folder.replace('_original', '_mediated')
        create_folder(label, PATH2)
        files = find_all_files(PATH + folder + "/")
        for fileName in files:
            source = folder.replace('_original', '')
            relabeling_projection_file(fileName, source)
            
```
